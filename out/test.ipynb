```python
import numpy as np

def normalize_array(arr):
    """
    Normalize an array to have values between 0 and 1.
    
    Args:
    arr (numpy.ndarray): Input array.

    Returns:
    numpy.ndarray: Normalized array.
    """
    min_val = np.min(arr)
    max_val = np.max(arr)
    normalized_arr = (arr - min_val) / (max_val - min_val)
    return normalized_arr

```
